```python
import tsanalysis.asdetect as ta
import numpy as np
```

```python
N = 103
n = 10

ix = ta.centered_segmentation(N,n)
a = np.arange(N)
print(ix)


for i in range(len(ix)-1):
    print('len :',len(a[ix[i]:ix[i+1]]))
    print('array :',a[ix[i]:ix[i+1]])
```

We can see here pretty good how the function is segmenting the given array. Each segment, including the first and the last, have the same length n. The function gives back the indices ix which mark the first index of each segment. Naturally, some of the datapoints do not fit into a segment with length n. These extra datapoints are truncated. The datapoints choosen to be truncated are at the beginning and at the end. In this case above the datapoints with indices [0,101,102] are choose. We can see in this exmple than in an uneven case, more datapoints are truncated at the end of the data array.
